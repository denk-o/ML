```python
import pandas as pd
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('data/arabica_data_cleaned.csv')
```

```python
data.columns
```

```python
data.corr()
```

```python
my_lasso = LassoCV()
y = data["Moisture"]
X = data[["Aroma", "Flavor", "Aftertaste", "Acidity", "Body", "Balance", "Uniformity", "Sweetness"]]
sc = StandardScaler()

x_tr, x_te, y_tr, y_te = train_test_split(X,y, random_state=88)

sc.fit(x_tr)
x_tr = sc.transform(x_tr)
x_te = sc.transform(x_te)
```

```python
my_lasso.fit(x_tr, y_tr)
```

```python
np.sqrt(np.mean((y_te - my_lasso.predict(x_te))**2))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.datasets import make_classification

x_t_reg, x_test, y_t_reg, y_test = train_test_split(X,y, random_state=0)
reg = RandomForestRegressor(max_depth=2, random_state=88)

sc.fit(x_t_reg)
x_t_reg = sc.transform(x_t_reg)
x_test = sc.transform(x_test)
reg.fit(x_t_reg,y_t_reg)
```

```python
np.sqrt(np.mean(y_test - reg.predict(x_test))**2)
```
